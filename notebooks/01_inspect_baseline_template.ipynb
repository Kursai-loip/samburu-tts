## üéôÔ∏è Step 3 ‚Äì Generate Test Audio (via Gradio UI)

```python
import gradio as gr
from transformers import AutoProcessor, AutoModelForTextToWaveform
import torch, soundfile as sf

# Load model and processor
model_id = "facebook/mms-tts-saq"
processor = AutoProcessor.from_pretrained(model_id)
model = AutoModelForTextToWaveform.from_pretrained(model_id).to("cuda")

# Define the synthesis function
def synthesize(text):
    inputs = processor(text=text, return_tensors="pt").to("cuda")
    with torch.no_grad():
        waveform = model(**inputs).waveform[0].cpu()
    sf.write("output.wav", waveform.numpy(), samplerate=16000)
    return "output.wav"

# Define the Gradio interface
ui = gr.Interface(
    fn=synthesize,
    inputs=gr.Textbox(label="Enter Samburu Text", placeholder="Type your Samburu sentence here..."),
    outputs=gr.Audio(label="Generated Speech"),
    title="Samburu TTS Baseline Test",
    description=(
        "This interface uses the Facebook MMS-TTS-SAQ checkpoint.\n"
        "Enter Samburu text below, click Generate, and listen to the synthesized audio.\n\n"
        "Use this to note where pronunciation deviates from natural Samburu tone or rhythm."
    ),
    allow_flagging="manual"
)

# Launch the app
ui.launch(debug=True)
```
